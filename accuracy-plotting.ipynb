```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define your data
data = {
    'Model': ['Naive Bayes', 'Logistic Regression', 'SVM', 'KNN', 'SNN', 'SNN + Random Forest'],
    'Accuracy': [75.8, 78.9, 78.5, 92.6, 93.1, 98.3],
    'Precision': [71, 76, 76, 87, 89, 97],
    'Recall': [87, 84, 84, 99, 98, 99.2],
    'F1-Score': [78, 80, 80, 93, 93, 98],
    'AUC': [83, 85, 85, 97, 97, 98]
}

df = pd.DataFrame(data)

# Set the style
sns.set(style="whitegrid")

# Set up the matplotlib figure
f, axes = plt.subplots(5, 1, figsize=(10, 15), sharex=True)

# Plot Accuracy
ax = sns.barplot(x='Accuracy', y='Model', data=df, palette='Blues', ax=axes[0])
axes[0].set_xlabel('Accuracy (%)')

# Plot Precision
ax = sns.barplot(x='Precision', y='Model', data=df, palette='Greens', ax=axes[1])
axes[1].set_xlabel('Precision (%)')

# Plot Recall
ax = sns.barplot(x='Recall', y='Model', data=df, palette='Reds', ax=axes[2])
axes[2].set_xlabel('Recall (%)')

# Plot F1-Score
ax = sns.barplot(x='F1-Score', y='Model', data=df, palette='Purples', ax=axes[3])
axes[3].set_xlabel('F1-Score (%)')

# Plot AUC
ax = sns.barplot(x='AUC', y='Model', data=df, palette='Oranges', ax=axes[4])
axes[4].set_xlabel('AUC (%)')

# Adding labels to the bars
for ax in axes:
    for p in ax.patches:
        ax.annotate(f'{p.get_width():.2f}%', (p.get_width() + 0.5, p.get_y() + p.get_height() / 2),
                    ha='center', va='center', xytext=(5, 0), textcoords='offset points')

# Adding labels
plt.ylabel('Models', fontsize=12)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define your data
data = {
    'Model': ['Naive Bayes', 'Logistic Regression', 'SVM', 'KNN', 'SNN', 'SNN + Random Forest'],
    'Accuracy': [75.8, 78.9, 78.5, 92.6, 93.1, 98.3],
    'Precision': [71, 76, 76, 87, 89, 97],
    'Recall': [87, 84, 84, 99, 98, 99.2],
    'F1-Score': [78, 80, 80, 93, 93, 98],
    'AUC': [83, 85, 85, 97, 97, 98]
}

df = pd.DataFrame(data)

# Set the style and color palette
sns.set(style="whitegrid")
palette = 'Blues'

# Set up the matplotlib figure
f, axes = plt.subplots(5, 1, figsize=(10, 15), sharex=True)

# Plot Accuracy
ax = sns.barplot(x='Accuracy', y='Model', data=df, palette=palette, ax=axes[0])
axes[0].set_xlabel('Accuracy (%)')

# Plot Precision
ax = sns.barplot(x='Precision', y='Model', data=df, palette=palette, ax=axes[1])
axes[1].set_xlabel('Precision (%)')

# Plot Recall
ax = sns.barplot(x='Recall', y='Model', data=df, palette=palette, ax=axes[2])
axes[2].set_xlabel('Recall (%)')

# Plot F1-Score
ax = sns.barplot(x='F1-Score', y='Model', data=df, palette=palette, ax=axes[3])
axes[3].set_xlabel('F1-Score (%)')

# Plot AUC
ax = sns.barplot(x='AUC', y='Model', data=df, palette=palette, ax=axes[4])
axes[4].set_xlabel('AUC (%)')

# Adding labels to the bars
for ax in axes:
    for p in ax.patches:
        ax.annotate(f'{p.get_width():.2f}%', (p.get_width() + 0.5, p.get_y() + p.get_height() / 2),
                    ha='center', va='center', xytext=(5, 0), textcoords='offset points')

# Adding labels
plt.ylabel('Models', fontsize=12)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
